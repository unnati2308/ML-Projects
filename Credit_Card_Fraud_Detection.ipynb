```python
#Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import warnings
```

This dataset is highly unbalanced


0--> normal transaction |
 1--> fraudulent transactions

```python
#seperating the data for analysis
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(fraud.shape)
```

Under-Sampling:

build a dataset containing similar distributions of normal transactions and fraudulent transactions


number of fraud transactions=492

```python
legit_sample=legit.sample(n=492)
```

```python
#concatenationg 2 dataframes
new=pd.concat([legit_sample,fraud], axis=0)
```

```python
new.head()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
#splitting the data into features and targets
X=new.drop(columns='Class', axis=1)
Y=new['Class']
```

```python
X.head()
```

```python
Y.head()
```

```python
#splitting the data into training data and test data
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
#model training
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train, y_train)
```

```python
# model evaluation
# accuracy score
x_train_pred=model.predict(x_train)
train_data_accuracy=accuracy_score(x_train_pred,y_train)
```

```python
print('Accuracy on training data: ', train_data_accuracy)
```

```python
x_test_pred=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_pred, y_test)
```

```python
print('Accuracy of testing data: ',test_data_accuracy)
```
